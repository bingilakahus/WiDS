```python
#importing all libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import torch
from scipy.io import loadmat
from torchvision import transforms
```

```python
# Load the MNIST dataset
mnist = loadmat("C:\\Users\\rohan\\OneDrive - Indian Institute of Technology Bombay\\IITB\\WiDS\\mnist-original.mat")

# Extract data and labels
data = mnist['data']
labels = mnist['label'][0]

# Convert to PyTorch tensors
data_tensor = torch.Tensor(data)
labels_tensor = torch.Tensor(labels)

# Creating a reshaped tensor which has the same dimensions as that of the original image
reshaped_data = data_tensor.view(28, 28, 70000)
```

```python
# Visualising the first 10 images
from matplotlib import pyplot
for i in range(9):  
    pyplot.subplot(331)
    pyplot.imshow(reshaped_data[:,:,i], cmap=pyplot.get_cmap('gray'))
    plt.axis('off')
    pyplot.show()
```

```python
# Data Preprocessing
# Flatten the images into 1D arrays
flattened_data = data_tensor.view(data_tensor.size(0), -1)

# Normalize pixel values to a range between 0 and 1
normalized_data = flattened_data / 255.0
```

```python
# Descriptive Statistics
# Label distribution and finding the array
label_distribution = pd.Series(labels_tensor.numpy().astype(int).flatten()).value_counts().sort_index()
print("Label Distribution:")
print(label_distribution)
```

```python
#getting a random image for a given specific label for each of the labels

for i in range(10):
    indices = torch.where(labels_tensor == i)
    flat_indices = torch.flatten(indices[0])
    random_index = torch.randint(0, flat_indices.size(0), (1,)).item()
    # Use the random index to get the corresponding value
    random_value = flat_indices[random_index]
    pyplot.subplot(331)
    pyplot.imshow(reshaped_data[:,:,random_value], cmap=pyplot.get_cmap('gray'))
    plt.axis('off')
    pyplot.show()
```

```python
# Visualising the first 10 images
from matplotlib import pyplot
import numpy as np
# Generate a list of 10 random numbers from 1 to 70,000
random_numbers = np.random.randint(1, 70001, size=10)

for i in random_numbers:  
    pyplot.subplot(331)
    pyplot.imshow(reshaped_data[:,:,i], cmap=pyplot.get_cmap('gray'))
    plt.axis('off')
    pyplot.show()
```

```python
# Plotting the bar graph
plt.bar(range(10), label_distribution)

# Adding labels and title
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Bar Graph of Tensor')

# Showing the plot
plt.show()
```

```python
# Choosing a random image number 40000 and seeing the original image
pyplot.subplot(331)
pyplot.imshow(reshaped_data[:,:,40000], cmap=pyplot.get_cmap('gray'))
plt.axis('off')
pyplot.show()
```

```python
import torchvision.transforms.functional as F

reshaped_data2 = reshaped_data.numpy()
img = reshaped_data2[:,:,40000]

# Convert the NumPy array to a PyTorch tensor
img_tensor = torch.from_numpy(img).unsqueeze(0).unsqueeze(0).float()

rotation_angle = 180  # Random angle used to adjust
rotate_transform = F.rotate(img_tensor, rotation_angle)

# Convert the rotated image tensor back to a NumPy array for visualization
rotated_img = rotate_transform.numpy().squeeze()
pyplot.subplot(331)
plt.imshow(rotated_img, cmap='gray')
plt.axis('off')
plt.show()
```
